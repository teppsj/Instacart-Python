```python
# Step 4: Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform()
#         and loc() functions.

# performing transform() function to create 'max_order' column

ords_prods_merg['max_order'] = ords_prods_merg.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# checking that the function worked

ords_prods_merg.head(100)
```

```python
# Creating loyalty flags - Loyal customer flag

ords_prods_merg.loc[ords_prods_merg['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Regular Customer flag

ords_prods_merg.loc[(ords_prods_merg['max_order'] <= 40) & (ords_prods_merg['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# New Customer flag

ords_prods_merg.loc[ords_prods_merg['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# checking that the flag worked

ords_prods_merg['loyalty_flag'].value_counts(dropna = False)
```

```python
# Step 5: The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three 
#         types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product 
#         prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to 
#         determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or 
#         new customers.

# Finding the mean of prices bought for each loyalty flag category

ords_prods_merg.groupby('loyalty_flag').agg({'prices': ['mean']})
```

On average, new customers spend about 3 dollars more per product than loyal customers, and 60 cents more than regular customers.

```python
# Step 6: The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by 
#         looking at the prices of the items people are buying. Create a spending flag for each user based on the average price 
#         across all their orders using the following criteria:
#   If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#   If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

# Aggregate the mean of price per user_id

ords_prods_merg.groupby('user_id').agg({'prices': ['mean']})
```

```python
# performing transform() function

ords_prods_merg['avg_price'] = ords_prods_merg.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking that variable column was created correctly

ords_prods_merg.head(50)
```

```python
# Creating spending flags

ords_prods_merg.loc[ords_prods_merg['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merg.loc[ords_prods_merg['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking the flag worked

ords_prods_merg['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merg.head(15)
```

```python
# Step 7: In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the 
#         same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order 
#         frequency flag that marks the regularity of a user’s ordering behavior according to the median in the 
#         “days_since_prior_order” column. The criteria for the flag should be as follows:
#    If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#    If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#    If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

# Aggregate the median of days_since_prior_order per user_id

ords_prods_merg.groupby('user_id').agg({'days_since_prior_order': ['median']})
```

```python
# performing transform() function

ords_prods_merg['order_frequency'] = ords_prods_merg.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking that variable column was created correctly

ords_prods_merg.head(100)
```

```python
# Creating order frequency flags

ords_prods_merg.loc[ords_prods_merg['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merg.loc[(ords_prods_merg['order_frequency'] <= 20) & (ords_prods_merg['order_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merg.loc[ords_prods_merg['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking the flag worked

ords_prods_merg['frequency_flag'].value_counts(dropna = False)
```
