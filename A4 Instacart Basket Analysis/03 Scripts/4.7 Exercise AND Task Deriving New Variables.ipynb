```python
# Creating a subset of the dataframe

df = df_ords_prods_merged[:1000000]
```

```python
# Creating price range flags

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply new fucntion

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Step 4: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at 
#         Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled 
#         “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called 
#         “busiest_period_of_day.”

# Checking the hours of day column to see details

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

I've chosen to divide the 24 hours up into thrids to make the different category conditions as a specific number or level was not specified.

```python
# Creating Conditions

result = []

for value in df_ords_prods_merged['order_hour_of_day']:
  if (value > 8) & (value <= 16):
    result.append("Most Orders")
  elif (value > 6) & (value <= 8):
    result.append("Average Orders")
  elif (value > 16) & (value <= 22):
    result.append("Average Orders")
  else:
    result.append("Fewest Orders")
```

```python
result
```

```python
# creating a new column

df_ords_prods_merged['busiest_period_of_day'] = result
```

```python
# Step 5: Print the frequency for this new column.

df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
